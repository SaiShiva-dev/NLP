```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/SMS.csv', encoding='latin-1')
df.head()
```

```python
message = df['Message_body']
label = df['Label']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
cv_matrix = cv.fit_transform(message)
cv_df = pd.DataFrame(cv_matrix.toarray(), columns=cv.get_feature_names_out())
cv_df.head()
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(cv_matrix, label)
```

```python
def predict_sms(sms):
  cv_sms = cv.transform([sms])
  return svc.predict(cv_sms)
```

```python
result = predict_sms('Rofl. Its true to its name')
print(result)
```

```python
result = predict_sms('Will ü b going to esplanade fr home?')
print(result)
```

By using TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer()
tf_matrix = tf.fit_transform(message)
tf_df = pd.DataFrame(tf_matrix.toarray(), columns=tf.get_feature_names_out())
tf_df.head()
```

```python
svc.fit(tf_matrix, label)
```

```python
def predict_sms_tf(sms):
  tf_sms = tf.transform([sms])
  return svc.predict(tf_sms)
```

```python
result = predict_sms_tf('Rofl. Its true to its name')
print(result)
```

```python
result = predict_sms_tf('Will ü b going to esplanade fr home?')
print(result)
```
